<h1><u>2. Performance analysis based on Gender.</u></h1>


---

```python
# Unique entries of Gender column
sp['gender'].unique()
```

```python
# Grouping by Gender
temp1=sp[sp.gender=='male']
temp2=sp[sp.gender=='female']
```

```python
# Gender wise Numerical analysis
sp.groupby(['gender']).agg(['min','median','max','mean'])
```

---

<h2>Females have a higher mean and median for total marks which indicates they have performed better than most males while males have performed better when only mathematics is considered. Also the minimum scores in all subjects have been scored by females indicating some of the girls have performed very poorly.</h2>

---

```python
# Number of males and females
sns.countplot(sp['gender'])
```

```python
# Total score distribution based on gender
plt.figure(figsize=(12,7))
sns.kdeplot(temp1['total'],label='male')
sns.kdeplot(temp2['total'],label='female')
plt.xlabel('total')
```

```python
# Math score distribution based on gender
plt.figure(figsize=(12,7))
sns.kdeplot(temp1['math score'],label='male')
sns.kdeplot(temp2['math score'],label='female')
plt.xlabel('math score')
```

```python
# Reading score distribution based on gender
plt.figure(figsize=(12,7))
sns.kdeplot(temp1['reading score'],label='male')
sns.kdeplot(temp2['reading score'],label='female')
plt.xlabel('reading score')
```

```python
# Writing score distribution based on gender
plt.figure(figsize=(12,7))
sns.kdeplot(temp1['writing score'],label='male')
sns.kdeplot(temp2['writing score'],label='female')
plt.xlabel('writing score')
```

---

<h2>Graphically it can be seen that females have larger peaks at higher marks(Except Maths) and their distribution is more tilted towards the higher marks region. Therefore it can be concluded that females have performed better than males.</h2>

<h1><u>3. Performance analysis based on Race/Ethnicity.</u></h1>

---

```python
# Unique entries of race/ethnicity column
sp['race/ethnicity'].unique()
```

```python
# Grouping by race/ethnicity
temp1=sp[sp['race/ethnicity']=='group A']
temp2=sp[sp['race/ethnicity']=='group B']
temp3=sp[sp['race/ethnicity']=='group C']
temp4=sp[sp['race/ethnicity']=='group D']
temp5=sp[sp['race/ethnicity']=='group E']         
```

```python
# Group wise Numerical analysis
sp.groupby(['race/ethnicity']).agg(['min','median','max','mean'])
```

---

<h2>Group E has the highest mean and median in all sections indicating that they have performed better than other groups. Similarly Group A has the lowest average and lowest median in all sections thus indicating poor performance. Overall order can be stated as E>D>C>B>A. Also the minimum scores in all sections have been scored by Group C students indicating some of the Group C students have performed very poorly.</h2>

---

```python
# Number of students in each group
sns.countplot(sp['race/ethnicity'])
```

```python
# Total score distribution based on race/ethnicity
plt.figure(figsize=(18,10))
sns.distplot(temp1['total'],kde=True,bins=10,label='group A',hist=False)
sns.distplot(temp2['total'],kde=True,bins=10,label='group B',hist=False)
sns.distplot(temp3['total'],kde=True,bins=10,label='group C',hist=False)
sns.distplot(temp4['total'],kde=True,bins=10,label='group D',hist=False)
sns.distplot(temp5['total'],kde=True,bins=10,label='group E',hist=False)
plt.legend()
```

```python
# Math score distribution based on race/ethnicity
plt.figure(figsize=(18,10))
sns.distplot(temp1['math score'],kde=True,bins=10,label='group A',hist=False)
sns.distplot(temp2['math score'],kde=True,bins=10,label='group B',hist=False)
sns.distplot(temp3['math score'],kde=True,bins=10,label='group C',hist=False)
sns.distplot(temp4['math score'],kde=True,bins=10,label='group D',hist=False)
sns.distplot(temp5['math score'],kde=True,bins=10,label='group E',hist=False)
plt.legend()
```

```python
# Reading score distribution based on race/ethnicity
plt.figure(figsize=(18,10))
sns.distplot(temp1['reading score'],kde=True,bins=10,label='group A',hist=False)
sns.distplot(temp2['reading score'],kde=True,bins=10,label='group B',hist=False)
sns.distplot(temp3['reading score'],kde=True,bins=10,label='group C',hist=False)
sns.distplot(temp4['reading score'],kde=True,bins=10,label='group D',hist=False)
sns.distplot(temp5['reading score'],kde=True,bins=10,label='group E',hist=False)
plt.legend()
```

```python
# Writing score distribution based on race/ethnicity
plt.figure(figsize=(18,10))
sns.distplot(temp1['writing score'],kde=True,bins=10,label='group A',hist=False)
sns.distplot(temp2['writing score'],kde=True,bins=10,label='group B',hist=False)
sns.distplot(temp3['writing score'],kde=True,bins=10,label='group C',hist=False)
sns.distplot(temp4['writing score'],kde=True,bins=10,label='group D',hist=False)
sns.distplot(temp5['writing score'],kde=True,bins=10,label='group E',hist=False)
plt.legend()
```

---

<h2>Graphically it can be seen that Group E students have larger peaks at higher marks and their distribution is more tilted towards the higher marks region and similar analysis can be seen in other groups with the final order as E>D>C>B>A</h2>

```python
# Math score distribution based on lunch type.
plt.figure(figsize=(12,7))
sns.distplot(temp1['math score'],kde=True,bins=10,label="standard",hist=False)
sns.distplot(temp2['math score'],kde=True,bins=10,label='free/reduced',hist=False)
plt.legend()
```

```python
# Reading score distribution based on lunch type.
plt.figure(figsize=(12,7))
sns.distplot(temp1['reading score'],kde=True,bins=10,label="standard",hist=False)
sns.distplot(temp2['reading score'],kde=True,bins=10,label='free/reduced',hist=False)
plt.legend()
```

```python
# Writing score distribution based on lunch type.
plt.figure(figsize=(12,7))
sns.distplot(temp1['writing score'],kde=True,bins=10,label="standard",hist=False)
sns.distplot(temp2['writing score'],kde=True,bins=10,label='free/reduced',hist=False)
plt.legend()
```

---

<h2>Graphically it can be seen that students with standard lunch have larger peaks at higher marks and their distribution is more tilted towards the higher marks region. Thus it can be concluded that students with standard lunch have fared better than subsidised lunch students</h2>
